```python
import folium
```

```python
# folium.Map() - 지도 생성
m = folium.Map(location=[37.547631, 126.942463], zoom_start=15, tiles="openstreetmap", width=800, height=600)  #tiles : 타일 스타일 종류 / width, height : 가로, 세로
m
```

```python
# folium.marker() - 마커 추가(popup, tooltip, icon)
folium.Marker(location=[37.547631, 126.942463], popup="팝업", tooltip="튤팁", icon=folium.Icon(color="black", icon="fa-solid fa-train-subway", prefix="fa-solid")).add_to(m)
m
```

```python
# foliu.ClickForMarker()
m.add_child(folium.ClickForMarker(popup="클릭"))
m
```

```python
#folium.LatLngPopup()
m.add_child(folium.LatLngPopup())
m
```

```python
# 실습. 마포구 택시 승차대 위치 표시하기
import pandas as pd
data = pd.read_csv("택시승차대 현황.csv", encoding='cp949')
taxi = data.loc[:, ["위치명", "위도", "경도"]]

for index, row in taxi.iterrows():
    folium.Marker(
        location=[row["위도"], row["경도"]], 
        popup = row["위치명"],
        icon = folium.Icon(color="red", icon="fa-solid fa-taxi", prefix="fa-solid")
    ).add_to(m)
m
```
