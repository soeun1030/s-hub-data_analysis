```python
from bs4 import BeautifulSoup

html_str = '''  #따옴표 : 여러 줄을 스트링으로 만들어줌(여러 줄을 한번에 저장할 때)
<html>
    <body>
        <div id="content">
            <ul class = 'industry'>
                <li>인공지능</li>
                <li>빅데이터</li>
                <li>스마트팩토리</li>
            </ul>
            <ul class = 'comlang'>
                <li>Python</li>
                <li>C++</li>
                <li>Javascript</li>
            </ul>
        </div>
    </body>
</html>
'''
soup = BeautifulSoup(html_str, "html.parser")  #html 파싱
```

```python
# beautifulsoup - find() 사용하기 - 처음 나오는 ul 태그로 찾기
first_ul = soup.find("ul")
print(first_ul)
print(first_ul.text)
print(first_ul.get_text())  #first_ul.text랑 같음
```

```python
# beautifulsoup - find() 사용하기 - 모든 요소를 li 태그로 찾기
all_list = first_ul.findAll("li")
print(all_list)
print(all_list[0])
print(all_list[0].text)

for li in all_list:
    print(li.text)
```

```python
# beautifulsoup - find() 사용하기 - class 선택자로 찾기
find_ul = soup.find("ul", class_="comlang")  #"ul" 빼도 동작
print(find_ul)
print(find_ul.text)
```

```python
# beautifulsoup - select() 사용하기 - select_one
f_ul = soup.select_one("ul.industry")
print(f_ul)
print(f_ul.text)
```

```python
# beautifulsoup - select() 사용하기 - select
all_li = f_ul.select("li")
print(all_li)
print(all_li[0])
print(all_li[0].text)
```

```python
all_li = soup.select("li")
print(all_li)
print(all_li[0])
print(all_li[0].text)
```

```python
all_li = soup.select("ul.industry > li")
print(all_li)
print(all_li[0])
print(all_li[0].text)
```

```python
idid = soup.select("#content")
print(idid)
print(idid[0].text)
```

```python
# 실습. 사용자에게 검색어를 입력 받아 검색된 뉴스의 제목과 링크 가져오기
import requests
from bs4 import BeautifulSoup
search = input("검색어를 입력하세요: ")
response = requests.get(f"https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query={search}")
html = response.text
soup = BeautifulSoup(html, "html.parser")

news = soup.select(".news_tit")

for item in news:
    title = item.text
    link = item["href"]
    print(f"{title} & {link}")
```

```python
# 실습1. 환율정보 크롤링하기
import requests
from bs4 import BeautifulSoup

response = requests.get(f"https://finance.naver.com/marketindex/")
html = response.text
soup = BeautifulSoup(html, "html.parser")

exchangeList = soup.select("ul.data_lst li")

for i in exchangeList[:4]:
    currency_name = i.select_one(".h_lst").text
    exchange_rate = i.select_one(".value").text
    print(f"{currency_name} {exchange_rate}")
```
