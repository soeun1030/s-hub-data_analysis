```python
import matplotlib
```

```python
# 기본 그래프 그리기
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])  #x 좌표 생략
plt.show()  #그래프 표시(굳이 안해도됨)
```

```python
plt.plot([0,1,2,3],[1,2,3,4])  #x,y 좌표 모두 설정
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
font1 = {"family" : "fantasy", "color" : "deeppink", "weight" : "normal", "size" : "xx-large"}
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("x", labelpad=30, loc="left", fontdict=font1)  #어떤 내용 쓸지 수정 가능 / left or right / labelpad 숫자 변경 가능, font 굳이 설정안해도됨
plt.ylabel("YY", labelpad=50, loc="top", fontdict=font1)  #top or bottom
plt.show()
```

```python
#  기본 그래프 그리기 - 범례 표시하기
line1, = plt.plot([1,2,3], label="label1")
line2, = plt.plot([2,4,6], label="label2")
line3, = plt.plot([3,6,9], label="label3")
# plt.legend(handles=[line1, line2], loc=(1,1))  #label 창 위치 : loc handless빼면 line1,2,3 뒤에 쉼표빼도 됨
plt.legend()  #바로 위에 줄 안적고(loc안하고) 얘만 적으면 항상 왼쪽 위에 위치함
plt.show()
```

```python
line1, = plt.plot([1,2,3], label="label1")
line2, = plt.plot([2,4,6], label="label2")
line3, = plt.plot([3,6,9], label="label3")
plt.legend(ncol=3, fontsize=14, frameon=True, shadow=True)  #ncol하면 위에 위치함 / frameon은 테두리, shadow는 그림자
plt.show()
```

```python
# 기본 그래프 그리기 - 타이틀 설정하기, 그리드 설정하기
line1, = plt.plot([1,2,3], label="label1")
line2, = plt.plot([2,4,6], label="label2")
line3, = plt.plot([3,6,9], label="label3")
plt.title("legend Test")  #타이틀 설정하기
plt.grid(True, axis="both")  #그리드 설정하기(x or y or both)
plt.show()
```

```python
# 막대 그래프 그리기
import numpy as np
x = np.arange(3)
y = [100,400,900]

plt.bar(x, y, color = ["r", "g", "b"], width = 0.7)
plt.plot(x,y)  #선그래프도 같이 그릴 수 있다

plt.xticks(x, ["2018", "2019", "2020"])
plt.yticks(y, ["a", "b", "c"])
plt.show()
```

```python
# 막대 그래프 그리기 - 수평막대그래프(barh 사용 - height으로 바꿔서 하기)
y = np.arange(3)
x = [100,400,900]

plt.barh(y, x, color = ["r", "g", "b"], height = 0.7)
plt.plot(x,y)  #선그래프도 같이 그릴 수 있다

plt.yticks(y, ["2018", "2019", "2020"])
plt.xticks(x, ["a", "b", "c"])
plt.show()
```

```python
# 산점도 그리기
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

area = (30 * np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x, y, area, colors, cmap="Spectral", alpha=0.5)
plt.show()
```

```python
x = [0,1,2]
y = [0,1,2]

area = [1000,1000,1000]
plt.plot(x,y)  #선그래프도 같이 그릴 수 있다

plt.scatter(x, y, area, alpha = 0.5)
plt.show()
```

```python
# 파이차트 그리기
ration = [24, 28, 12, 36]
labels = ["A", "B", "C", "D"]
wedgepros = {"width" : 0.7, "edgecolor" : "k"}

plt.pie(ration, labels = labels, wedgeprops=wedgepros, autopct="%.1f%%", explode=[0,0.1,0,0.1], counterclock = False, startangle = 90)
```

```python
# 실습1. 아래와 같은 그래프를 그려보세요.
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]

plt.plot(months, sales_2019, label = "2019")
plt.plot(months, sales_2020, label = "2020")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.title("Monthly Sales Comparison(2019-2020)")
```

```python
# 실습2. 아래와 같은 그래프를 그려보세요.
##나의 방법
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]
y = np.arange(0,51,10)

plt.bar(categories, data)
plt.xticks(categories, rotation = 45)
plt.yticks(y)
plt.xlabel("Categories")
plt.ylabel("Values")
plt.grid(True, axis="both")
plt.title("Bar Chart")

##방법2
# x = np.arange(5)
# plt.bar(x,data)
# plt.xticks(x, categories, rotation=45)
# plt.title("Bar Chart")
# plt.xlabel("Categories")
# plt.ylabel("Values")
# plt.grid(True)
# plt.ylim(0,50)

```

```python
# 실습3. 아래와 같은 그래프를 그려보세요.
labels = ["Apple", "Banana", "Mango", "Blueberry"]
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

plt.pie(sizes, labels = labels, autopct="%.1f%%", explode=explode, startangle=90, shadow=True)
```

```python
# 실습4. sin & cos 함수 그리기
x = np.linspace(0, 10, 100)  #0에서 10사이를 100개로 나눈 것

y_sin = np.sin(x)
y_cos = np.cos(x)
plt.plot(x, y_sin, color = "blue", label="sin(x)")
plt.plot(x, y_cos, color = "red", label="cos(x)")
plt.legend()
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Sine and Cosine Functions")
```
